```python
import os
import time

def run_score_calculator():
    """리스트를 활용한 점수 입력 및 통계(평균/최고/최저) 계산"""
    print("\n--- [2] 학생 점수 통계 프로그램 ---")
    scores = []

    while True:
        user_input = input(
            "점수를 입력하세요 (종료하려면 'q' 입력): "
        ).strip()
        if user_input.lower() == "q":
            break

        try:
            score = float(user_input)
            if 0 <= score <= 100:
                scores.append(score)
            else:
                print("점수는 0점에서 100점 사이여야 합니다.")
        except ValueError:
            print("올바른 숫자를 입력해주세요.")

    if not scores:
        print("입력된 점수가 없습니다.")
        return

    avg_score = sum(scores) / len(scores)
    max_score = max(scores)
    min_score = min(scores)

    print("\n[ 통계 결과 ]")
    print(f"입력된 학생 수: {len(scores)}명")
    print(f"평균 점수: {avg_score:.2f}점")
    print(f"최고 점수: {max_score}점")
    print(f"최저 점수: {min_score}점")
run_score_calculator()
```
