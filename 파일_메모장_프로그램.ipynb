```python
import os


def run_notepad():
    """텍스트 파일 저장 및 불러오기 (메모장 전용 무한 루프)"""
    filename = "memo.txt"

    while True:
        print("\n" + "=" * 30)
        print("      간단한 파일 메모장")
        print("=" * 30)
        print("1: 메모 새로 쓰기 (이어쓰기)")
        print("2: 저장된 메모 읽기")
        print("3: 프로그램 완전히 종료")
        print("=" * 30)

        mode = input("원하는 작업 번호를 입력하세요: ").strip()

        if mode == "1":
            print("\n[메모 쓰기] 저장할 내용을 입력하세요. (취소하려면 빈 줄에서 엔터)")
            content = input("> ")

            # 아무것도 입력하지 않고 엔터치면 취소
            if not content.strip():
                print("입력이 취소되었습니다.")
                continue

            # 파일에 이어서 저장
            with open(filename, "a", encoding="utf-8") as f:
                f.write(content + "\n")
            print(f"🎉 '{filename}'에 성공적으로 저장되었습니다.")

        elif mode == "2":
            # 파일이 존재하는지 먼저 확인 (에러 방지)
            if not os.path.exists(filename):
                print("\n⚠️ 저장된 메모 파일이 없습니다. 먼저 메모를 작성해주세요.")
                continue

            print(f"\n--- [{filename}] 저장된 내용 ---")
            with open(filename, "r", encoding="utf-8") as f:
                print(f.read().rstrip())
            print("-" * 30)

        elif mode == "3":
            print("메모장 프로그램을 완전히 종료합니다. 이용해 주셔서 감사합니다!")
            break  # 루프를 탈출하여 프로그램 종료

        else:
            print("❌ 잘못된 입력입니다. 1, 2, 3번 중에서 선택해주세요.")


# 프로그램 실행
if __name__ == "__main__":
    run_notepad()
```
