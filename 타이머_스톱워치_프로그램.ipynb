```python
import time


def run_timer():
    """time 모듈을 사용한 초 단위 타이머 (무한 반복 기능 추가)"""

    while True:
        print("\n--- [5] 타이머 프로그램 ---")
        print("종료하고 싶다면 'q'를 입력하세요.")

        user_input = input("설정할 시간(초)을 입력하세요: ").strip()

        # 종료 조건 처리
        if user_input.lower() == "q":
            print("타이머 프로그램을 완전히 종료합니다.")
            break

        try:
            seconds = int(user_input)
            if seconds <= 0:
                print("⚠️ 1초 이상의 시간을 입력해주세요.")
                continue

            print("⏰ 타이머를 시작합니다.")
            for i in range(seconds, 0, -1):
                # end='\r'를 사용해 한 줄에서 숫자가 카운트다운되게 합니다.
                print(f"남은 시간: {i}초 ", end="\r")
                time.sleep(1)

            print("\n🔔 시간이 종료되었습니다!")

        except ValueError:
            print("❌ 올바른 정수 값을 입력하거나 'q'를 입력해주세요.")


# 프로그램 실행
if __name__ == "__main__":
    run_timer()
```
