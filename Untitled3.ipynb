```python
def add(a, b):
    return a + b


def subtract(a, b):
    return a - b


def multiply(a, b):
    return a * b


def divide(a, b):
    if b == 0:
        raise ZeroDivisionError("0으로 나눌 수 없습니다.")
    return a / b


def run_calculator():
    """계산기 실행 함수 (반복 기능 추가)"""
    print("\n--- [3] 간단한 계산기 ---")

    # 무한 루프를 돌려서 계산이 끝나도 계속 입력창이 나오게 합니다.
    while True:
        print("\n계산기를 종료하려면 연산자에 'q'를 입력하세요.")
        try:
            # 1. 첫 번째 숫자 입력
            num1_input = input("첫 번째 숫자 입력: ").strip()
            if num1_input.lower() == "q":
                print("계산기를 종료합니다.")
                break
            num1 = float(num1_input)

            # 2. 연산자 입력
            operator = input("연산자 입력 (+, -, *, /): ").strip()
            if operator.lower() == "q":
                print("계산기를 종료합니다.")
                break

            # 3. 두 번째 숫자 입력
            num2_input = input("두 번째 숫자 입력: ").strip()
            if num2_input.lower() == "q":
                print("계산기를 종료합니다.")
                break
            num2 = float(num2_input)

            # 연산 수행
            if operator == "+":
                print(f"결과: {add(num1, num2)}")
            elif operator == "-":
                print(f"결과: {subtract(num1, num2)}")
            elif operator == "*":
                print(f"결과: {multiply(num1, num2)}")
            elif operator == "/":
                print(f"결과: {divide(num1, num2)}")
            else:
                print("지원하지 않는 연산자입니다. 다시 시도해주세요.")

        except ValueError:
            print("숫자 입력이 올바르지 않습니다. 처음부터 다시 입력해주세요.")
        except ZeroDivisionError as e:
            print(e)


if __name__ == "__main__":
    run_calculator()
```
